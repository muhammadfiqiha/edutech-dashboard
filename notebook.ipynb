```python
def numerical_dis_plot(features, df, segment_feature=None, showfliers=True):
    fig, ax = plt.subplots(len(features), 1,figsize=(15,30))
    for i, feature in enumerate(features):
        if segment_feature:
            sns.boxplot(y=segment_feature, x=feature, data=df, ax=ax[i], showfliers=showfliers)
            ax[i].set_ylabel(None)
        else:
            sns.boxplot(x=feature, data=df, ax=ax[i], showfliers=showfliers)
    plt.tight_layout()
    plt.show()

num_columns = [col for col in cleaned_df.columns if col not in cat_cols]
numerical_dis_plot(
    features=num_columns,
    df=cleaned_df
)
```

```python
def categorical_plot(features, df, segment_feature=None, fig_size=(10, 4)):
    binary_meanings = {
        'Displaced': '0 = No, 1 = Yes',
        'Educational_special_needs': '0 = No, 1 = Yes',
        'Debtor': '0 = No, 1 = Yes',
        'Tuition_fees_up_to_date': '0 = No, 1 = Yes',
        'Gender': '0 = Female, 1 = Male',
        'Scholarship_holder': '0 = No, 1 = Yes',
        'Daytime_evening_attendance': '0 = Evening, 1 = Daytime',
        'International': '0 = No, 1 = Yes'
    }

    n = len(features)
    fig, ax = plt.subplots(n, 1, figsize=(fig_size[0], fig_size[1]*n))

    if n == 1:
        ax = [ax]

    for i, feature in enumerate(features):

      cat_order = None

      if features == multi_obj_wo_status:
        dropout_counts = (
            df[df[segment_feature] == 'Dropout'][feature]
            .value_counts()
            .sort_values(ascending=False)
            )
        cat_order = dropout_counts.index.tolist()

      if segment_feature:
          sns.countplot(data=df, x=feature, hue=segment_feature, ax=ax[i], order=cat_order)
          ax[i].legend(title=segment_feature)
      else:
          sns.countplot(data=df, y=feature, ax=ax[i])

      title = f"{feature}"
      if feature in binary_meanings:
          title += f" ({binary_meanings[feature]})"
      ax[i].set_title(title)
      ax[i].tick_params(axis='x', rotation=45)

    plt.tight_layout()
    plt.show()


# see distribution of Status based on categorical column (binary value)
categorical_plot(
    features=binary_obj_cols,
    df=cleaned_df,
    segment_feature="Status"
)
```

Because the data is imbalance, we'll see based on proportion/percentage

```python
def plot_categorical_proportions(df, features, segment_feature, fig_size=(10, 4)):
    binary_meanings = {
        'Displaced': '0 = No, 1 = Yes',
        'Educational_special_needs': '0 = No, 1 = Yes',
        'Debtor': '0 = No, 1 = Yes',
        'Tuition_fees_up_to_date': '0 = No, 1 = Yes',
        'Gender': '0 = Female, 1 = Male',
        'Scholarship_holder': '0 = No, 1 = Yes',
        'Daytime_evening_attendance': '0 = Evening, 1 = Daytime',
        'International': '0 = No, 1 = Yes'
    }

    hue_order = ['Dropout', 'Graduate', 'Enrolled']

    n = len(features)
    fig, ax = plt.subplots(n, 1, figsize=(fig_size[0], fig_size[1]*n))

    if n == 1:
        ax = [ax]

    for i, feature in enumerate(features):
        prop_df = (
            df.groupby(feature, observed=True)[segment_feature]
            .value_counts(normalize=True)
            .rename("Proportion")
            .reset_index()
        )

        if features == multi_obj_wo_status:
          # Urutkan kategorinya berdasarkan dropout rate tertinggi
          dropout_rates = (
              prop_df[prop_df[segment_feature] == 'Dropout']
              .sort_values('Proportion', ascending=False)
          )
          category_order = dropout_rates[feature].tolist()
        else:
          category_order = sorted(df[feature].dropna().unique())

        sns.barplot(
            data=prop_df,
            x=feature,
            y='Proportion',
            hue=segment_feature,
            hue_order=hue_order,
            order=category_order,
            ax=ax[i]
        )

        title = f"{feature}"
        if feature in binary_meanings:
            title += f" ({binary_meanings[feature]})"
        ax[i].set_title(title)
        ax[i].tick_params(axis='x', rotation=45)
        ax[i].legend(title=segment_feature)

    plt.tight_layout()
    plt.show()


    plt.tight_layout()
    plt.show()


plot_categorical_proportions(
    df=cleaned_df,
    features=binary_obj_cols,
    segment_feature="Status"
)
```

### Encoding & Scaling

```python
X_train = resampled_train_df.drop(columns="Status", axis=1)
y_train = resampled_train_df["Status"]

X_test = test_df.drop(columns="Status", axis=1)
y_test = test_df["Status"]
```

```python
from sklearn.preprocessing import LabelEncoder
from sklearn.preprocessing import MinMaxScaler
import joblib
import os

os.makedirs("model", exist_ok=True)

def scaling(features, df, df_test=None):
    if df_test is not None:
        df = df.copy()
        df_test = df_test.copy()
        for feature in features:
            scaler = MinMaxScaler()
            X = np.asanyarray(df[feature])
            X = X.reshape(-1,1)
            scaler.fit(X)
            df["{}".format(feature)] = scaler.transform(X)
            joblib.dump(scaler, "model/scaler_{}.joblib".format(feature))

            X_test = np.asanyarray(df_test[feature])
            X_test = X_test.reshape(-1,1)
            df_test["{}".format(feature)] = scaler.transform(X_test)
        return df, df_test
    else:
        df = df.copy()
        for feature in features:
            scaler = MinMaxScaler()
            X = np.asanyarray(df[feature])
            X = X.reshape(-1,1)
            scaler.fit(X)
            df["{}".format(feature)] = scaler.transform(X)
            joblib.dump(scaler, "model/scaler_{}.joblib".format(feature))
        return df

def encoding(features, df, df_test=None):
    if df_test is not None:
        df = df.copy()
        df_test = df_test.copy()
        for feature in features:
            encoder = LabelEncoder()

            # Gabungkan nilai train dan test → agar encoder tahu semua kategori
            combined = pd.concat([df[feature], df_test[feature]]).astype(str)
            encoder.fit(combined)

            df[feature] = encoder.transform(df[feature].astype(str))
            df_test[feature] = encoder.transform(df_test[feature].astype(str))

            joblib.dump(encoder, f"model/encoder_{feature}.joblib")

        return df, df_test
    else:
        df = df.copy()
        for feature in features:
            encoder = LabelEncoder()
            df[feature] = df[feature].astype(str)
            encoder.fit(df[feature])
            df[feature] = encoder.transform(df[feature])
            joblib.dump(encoder, f"model/encoder_{feature}.joblib")
        return df
```

Later, those PCA group will be merged with other numeric features who arent included in those group.

```python
pca_numerical_columns_1 = [
    'Curricular_units_1st_sem_credited',
    'Curricular_units_1st_sem_enrolled',
    'Curricular_units_1st_sem_evaluations',
    'Curricular_units_1st_sem_approved',
    'Curricular_units_1st_sem_grade',
    'Curricular_units_1st_sem_without_evaluations',
]

pca_numerical_columns_2 = [
    'Curricular_units_2nd_sem_credited',
    'Curricular_units_2nd_sem_enrolled',
    'Curricular_units_2nd_sem_evaluations',
    'Curricular_units_2nd_sem_approved',
    'Curricular_units_2nd_sem_grade',
    'Curricular_units_2nd_sem_without_evaluations',
]
```

```python
train_pca_df = new_train_df.copy().reset_index(drop=True)
test_pca_df = new_test_df.copy().reset_index(drop=True)
```

```python
from sklearn.decomposition import PCA

# see scree plot from the first group

pca = PCA(n_components=len(pca_numerical_columns_1), random_state=123)
pca.fit(train_pca_df[pca_numerical_columns_1])
princ_comp = pca.transform(train_pca_df[pca_numerical_columns_1])

var_exp = pca.explained_variance_ratio_.round(3)
cum_var_exp = np.cumsum(var_exp)

plt.bar(range(len(pca_numerical_columns_1)), var_exp, alpha=0.5, align='center', label='individual explained variance')
plt.step(range(len(pca_numerical_columns_1)), cum_var_exp, where='mid', label='cumulative explained variance')
plt.ylabel('Explained variance ratio')
plt.xlabel('Principal component index')
plt.legend(loc='best')
plt.show()
```

Based on above, we will pick 2 components from the first group

```python
# first group PCA
pca_1 = PCA(n_components=2, random_state=123)
pca_1.fit(train_pca_df[pca_numerical_columns_1])
joblib.dump(pca_1, "model/pca_{}.joblib".format(1))
princ_comp_1 = pca_1.transform(train_pca_df[pca_numerical_columns_1])
train_pca_df[["pc1_1", "pc1_2"]] = pd.DataFrame(princ_comp_1, columns=["pc1_1", "pc1_2"])
train_pca_df.drop(columns=pca_numerical_columns_1, axis=1, inplace=True)
train_pca_df.head()
```

```python
# analyse scree plot from 2nd group

pca = PCA(n_components=len(pca_numerical_columns_2), random_state=123)
pca.fit(train_pca_df[pca_numerical_columns_2])
princ_comp = pca.transform(train_pca_df[pca_numerical_columns_2])

var_exp = pca.explained_variance_ratio_.round(3)
cum_var_exp = np.cumsum(var_exp)

plt.bar(range(len(pca_numerical_columns_2)), var_exp, alpha=0.5, align='center', label='individual explained variance')
plt.step(range(len(pca_numerical_columns_2)), cum_var_exp, where='mid', label='cumulative explained variance')
plt.ylabel('Explained variance ratio')
plt.xlabel('Principal component index')
plt.legend(loc='best')
plt.show()
```

Based on above, we will pick 2 components

```python
pca_2 = PCA(n_components=2, random_state=123)
pca_2.fit(train_pca_df[pca_numerical_columns_2])
joblib.dump(pca_2, "model/pca_{}.joblib".format(2))
princ_comp_2 = pca_2.transform(train_pca_df[pca_numerical_columns_2])
train_pca_df[["pc2_1", "pc2_2"]] = pd.DataFrame(princ_comp_2, columns=["pc2_1", "pc2_2"])
train_pca_df.drop(columns=pca_numerical_columns_2, axis=1, inplace=True)
train_pca_df.head()
```

```python
from sklearn.metrics import classification_report, confusion_matrix

def evaluating(y_pred, y_true):
    '''Evaluasi model'''
    labels=['Dropout', 'Enrolled', 'Graduate']

    print(classification_report(y_pred=y_pred, y_true=y_true))

    cnf_matrix = confusion_matrix(y_pred=y_pred, y_true=y_true, labels=labels)
    confusion_matrix_df = pd.DataFrame(cnf_matrix, labels, labels)
    sns.heatmap(confusion_matrix_df, annot=True, annot_kws={'size': 14}, fmt='d', cmap='YlGnBu')
    plt.ylabel('True label', fontsize=15)
    plt.xlabel('Predicted label', fontsize=15)
    plt.show()

    return confusion_matrix_df
```
